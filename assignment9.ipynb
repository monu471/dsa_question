Given an integerÂ `n`, returnÂ *`true`Â if it is a power of two. Otherwise, returnÂ `false`*.

An integerÂ `n`Â is a power of two, if there exists an integerÂ `x`Â such thatÂ `n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

</aside>

```python
def isPowerOfTwo(n):
    if n <= 0:
        return False

    # Check if only one bit is set to 1
    return (n & (n - 1)) == 0
```

****Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

</aside>

```python
def factorial(N):
    result = 1
    for i in range(1, N + 1):
        result *= i
    return result
```

Given an array of integersÂ **arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

</aside>

```python
def findMax(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        return max(arr[0], findMax(arr[1:]))
```

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = â€œABCâ€*

***Output:***

*â€œABCâ€, â€œACBâ€, â€œBACâ€, â€œBCAâ€, â€œCBAâ€, â€œCABâ€*

**Example 2:**

***Input:***

*S = â€œXYâ€*

***Output:***

*â€œXYâ€, â€œYXâ€*

</aside>

```python
def permute(s, left, right):
    if left == right:
        print("".join(s))
    else:
        for i in range(left, right + 1):
            s[left], s[i] = s[i], s[left]  # Swap characters
            permute(s, left + 1, right)    # Recursive call
            s[left], s[i] = s[i], s[left]  # Backtrack

def generatePermutations(s):
    n = len(s)
    permute(list(s), 0, n - 1)
```

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

</aside>

```python
def findProduct(arr):
    product = 1
    for num in arr:
        product *= num
    return product
```
