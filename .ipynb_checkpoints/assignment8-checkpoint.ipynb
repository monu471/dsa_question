ðŸ’¡ **Question 1**

Given two strings s1 and s2, return *the lowest **ASCII** sum of deleted characters to make two strings equal*.

**Example 1:**

**Input:** s1 = "sea", s2 = "eat"

**Output:** 231

**Explanation:** Deleting "s" from "sea" adds the ASCII value of "s" (115) to the sum.

Deleting "t" from "eat" adds 116 to the sum.

At the end, both strings are equal, and 115 + 116 = 231 is the minimum sum possible to achieve this

</aside>

```python
def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)

    # Create a table to store the lengths of the LCS
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Calculate the lengths of the LCS
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + ord(s1[i - 1])
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # Calculate the sum of ASCII values of deleted characters
    sum_s1 = sum(ord(c) for c in s1)
    sum_s2 = sum(ord(c) for c in s2)
    return sum_s1 + sum_s2 - 2 * dp[m][n]
```

ðŸ’¡ **Question 2**

Given a string s containing only three types of characters: '(', ')' and '*', return true *if* s *is **valid***.

The following rules define a **valid** string:

- Any left parenthesis '(' must have a corresponding right parenthesis ')'.
- Any right parenthesis ')' must have a corresponding left parenthesis '('.
- Left parenthesis '(' must go before the corresponding right parenthesis ')'.
- '*' could be treated as a single right parenthesis ')' or a single left parenthesis '(' or an empty string "".

**Example 1:**

**Input:** s = "()"

**Output:**

true

</aside>

```python
def isValid(s):
    stack = []

    for char in s:
        if char in '(*':
            stack.append(char)
        elif char == ')':
            if stack and stack[-1] != '(':
                stack.pop()
            elif stack and stack[-1] == '(':
                stack.pop()
            else:
                stack.append(char)

    return len(stack) == 0
```

ðŸ’¡ **Question 3**

Given two strings word1 and word2, return *the minimum number of **steps** required to make* word1 *and* word2 *the same*.

In one **step**, you can1 delete exactly one character in either string.

**Example 1:**

**Input:** word1 = "sea", word2 = "eat"

**Output:** 2

**Explanation:** You need one step to make "sea" to "ea" and another step to make "eat" to "ea".

</aside>

```python
def minStepsToDelete(word1, word2):
    m, n = len(word1), len(word2)

    # Create a table to store the lengths of the LCS
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Calculate the lengths of the LCS
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # Calculate the minimum number of steps required
    return m + n - 2 * dp[m][n]
```

ðŸ’¡ **Question 4**

You need to construct a binary tree from a string consisting of parenthesis and integers.

The whole input represents a binary tree. It contains an integer followed by zero, one or two pairs of parenthesis. The integer represents the root's value and a pair of parenthesis contains a child binary tree with the same structure.
You always start to construct the **left** child node of the parent first if it exists.

</aside>

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def constructBinaryTree(s):
    if not s:
        return None

    # Find the first occurrence of "("
    left_parenthesis_index = s.find("(")

    # If "(" is not found, the entire string is a single node
    if left_parenthesis_index == -1:
        return TreeNode(int(s))

    # Extract the value of the root node
    val = int(s[:left_parenthesis_index])

    # Find the index of the corresponding closing ")"
    right_parenthesis_index = findClosingParenthesis(s, left_parenthesis_index)

    # Construct the left child subtree
    left_subtree = constructBinaryTree(s[left_parenthesis_index + 1: right_parenthesis_index])

    # Construct the right child subtree if it exists
    right_subtree = None
    if right_parenthesis_index < len(s) - 1:
        right_subtree = constructBinaryTree(s[right_parenthesis_index + 2: -1])

    # Create the root node with its left and right child subtrees
    root = TreeNode(val, left_subtree, right_subtree)
    return root

def findClosingParenthesis(s, opening_index):
    count = 0
    for i in range(opening_index, len(s)):
        if s[i] == "(":
            count += 1
        elif s[i] == ")":
            count -= 1
            if count == 0:
                return i
    return -1
```

ðŸ’¡ **Question 5**

Given an array of characters chars, compress it using the following algorithm:

Begin with an empty string s. For each group of **consecutive repeating characters** in chars:

- If the group's length is 1, append the character to s.
- Otherwise, append the character followed by the group's length.

The compressed string s **should not be returned separately**, but instead, be stored **in the input character array chars**. Note that group lengths that are 10 or longer will be split into multiple characters in chars.

After you are done **modifying the input array,** return *the new length of the array*.

You must write an algorithm that uses only constant extra space.

**Example 1:**

**Input:** chars = ["a","a","b","b","c","c","c"]

**Output:** Return 6, and the first 6 characters of the input array should be: ["a","2","b","2","c","3"]

**Explanation:**

The groups are "aa", "bb", and "ccc". This compresses to "a2b2c3".

</aside>

```python
def compress(chars):
    n = len(chars)
    if n == 0:
        return 0

    count = 1  # Count of consecutive repeating characters
    write_index = 0  # Position to write the compressed characters

    for i in range(1, n):
        if chars[i] == chars[i - 1]:
            count += 1
        else:
            chars[write_index] = chars[i - 1]
            write_index += 1

            if count > 1:
                count_str = str(count)
                for j in range(len(count_str)):
                    chars[write_index] = count_str[j]
                    write_index += 1

            count = 1

    # Write the last character and its count if applicable
    chars[write_index] = chars[n - 1]
    write_index += 1

    if count > 1:
        count_str = str(count)
        for j in range(len(count_str)):
            chars[write_index] = count_str[j]
            write_index += 1

    return write_index
```

ðŸ’¡ **Question 6**

Given two strings s and p, return *an array of all the start indices of* p*'s anagrams in* s. You may return the answer in **any order**.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**

**Input:** s = "cbaebabacd", p = "abc"

**Output:** [0,6]

**Explanation:**

The substring with start index = 0 is "cba", which is an anagram of "abc".

The substring with start index = 6 is "bac", which is an anagram of "abc".

</aside>

```python
from collections import Counter

def findAnagrams(s, p):
    result = []
    len_s, len_p = len(s), len(p)

    if len_s < len_p:
        return result

    p_freq = Counter(p)
    window_freq = Counter(s[:len_p])

    if window_freq == p_freq:
        result.append(0)

    for i in range(len_p, len_s):
        left_char = s[i - len_p]
        right_char = s[i]

        window_freq[left_char] -= 1
        if window_freq[left_char] == 0:
            del window_freq[left_char]

        window_freq[right_char] += 1

        if window_freq == p_freq:
            result.append(i - len_p + 1)

    return result
```

ðŸ’¡ **Question 7**

Given an encoded string, return its decoded string.

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

The test cases are generated so that the length of the output will never exceed 105.

**Example 1:**

**Input:** s = "3[a]2[bc]"

**Output:** "aaabcbc"

</aside>

```python
def decodeString(s):
    stack = []
    count = 0
    decoded_string = ''

    for char in s:
        if char.isdigit():
            count = count * 10 + int(char)
        elif char == '[':
            stack.append((count, decoded_string))
            count = 0
            decoded_string = ''
        elif char == ']':
            prev_count, prev_decoded_string = stack.pop()
            decoded_string = prev_decoded_string + prev_count * decoded_string
        else:
            decoded_string += char

    return decoded_string
```

ðŸ’¡ **Question 8**

Given two strings s and goal, return true *if you can swap two letters in* s *so the result is equal to* goal*, otherwise, return* false*.*

Swapping letters is defined as taking two indices i and j (0-indexed) such that i != j and swapping the characters at s[i] and s[j].

- For example, swapping at indices 0 and 2 in "abcd" results in "cbad".

**Example 1:**

**Input:** s = "ab", goal = "ba"

**Output:** true

**Explanation:** You can swap s[0] = 'a' and s[1] = 'b' to get "ba", which is equal to goal.

</aside>

```python
def buddyStrings(s, goal):
    if len(s) != len(goal):
        return False

    if s == goal:
        # Check if there are at least two repeated characters in s
        char_count = [0] * 26
        for char in s:
            char_count[ord(char) - ord('a')] += 1
            if char_count[ord(char) - ord('a')] >= 2:
                return True
        return False

    mismatched_indices = []
    for i in range(len(s)):
        if s[i] != goal[i]:
            mismatched_indices.append(i)

    if len(mismatched_indices) != 2:
        return False

    i, j = mismatched_indices
    return s[i] == goal[j] and s[j] == goal[i]
```
